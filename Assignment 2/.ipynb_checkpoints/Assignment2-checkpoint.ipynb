```python
class NumpyClassifier():
    """Common methods to all numpy classifiers --- if any"""
    
    def accuracy(self,X_test, y_test, **kwargs):
        """
        Predicts and calculate accuracies for LinReg and LogReg.
        
            args:
                self(object)
                X_test (numpy ndarray): Points to predict
                y_test (numpy array): Labels to X_test
        """
        pred = self.predict(X_test, **kwargs)
        if len(pred.shape) > 1:
            pred = pred[:,0]
            
        return sum(pred==y_test)/len(pred)
```

```python
# Class of Linear Regression
class LinReg(NumpyClassifier):
    
    def __init__(self):
        """
        init of the class LinReg.
        
            param:
                theta (numpy ndarray): the 'weights'.
        """
        self.theta = None

    def fit(self, X_train, t_train, gamma = 0.1, diff=0.001):
        """
        training on the data given.
        
        The fit function will stop when the improvement is less than 'diff' 
        
            Args:
                X_train (numpy ndarray): matrix containing points.
                t_train (numpy array): validation-list.
                gamma (float): learning rate.
                diff (float): a limit to when the fit-function should stop. Compare 'improvement' to diff.
        """
        
        (k, m) = X_train.shape
        X_train = add_bias(X_train)
        
        self.theta = theta = np.zeros(m+1)
        
        #current mse
        current = mse(X_train, t_train, theta)
        run_counter = 0
        while True:
            #MeanSquareError
            theta -= gamma / k *  X_train.T @ (X_train @ theta - t_train)
            new_mse = mse(X_train, t_train, theta)
            # Difference in mse
            diff_curr_new = current - new_mse
            
            run_counter += 1
            if diff_curr_new > diff:
                current = new_mse
            else:
                break
        
        print(f"Total number of runs with diff as {diff}: {run_counter} runs")
            
    def predict(self, x, threshold=0.5):
        """
        Predicting the x by taking a 'forward-move'.
        
        Taking the dot-product to calculate the predictions.
        
            args:
                x (numpy ndarray): Values/points to test.
                threshold (float): Threshold if it should 'fire'.
                
            return:
                score (numpy ndarray)
        """
        z = add_bias(x)
        score = z @ self.theta
        return score>threshold
    
```

```python
def add_bias(X, spec_bias=1):
    """
    Adding a bias of 1 in position 0 in X.
    
        args:
            X (numpy ndarray):  Array to add bias to.
            
        return (numpy ndarray): Array with the added bias.
    """
    # Put bias in position 0
    sh = X.shape
    if len(sh) == 1:
        #X is a vector
        return np.concatenate([np.array([spec_bias]), X])
    else:
        # X is a matrix
        m = sh[0]
        bias = np.full((m,1), spec_bias) # Makes a m*1 matrix of 1-s
        return np.concatenate([bias, X], axis  = 1)

    
def mse(X, t, theta):
    """
    mean square error calculation.
    
        args:
            X (numpy ndarray): matrix containing points
            t (numpy array): array containing the classes
            theta (numpy-array): calculated weights
            
        return:
            error (float): mse
            
    """
    #y_estimate is prediction. Using dot product.
    y_estimate = X.dot(theta)
    error = np.sum((t - y_estimate)**2)/X.shape[0]
    
    return error
    
```

```python
# Function for fit and accuracy which can be used by multiple regression-functions.

def fit_and_accuracy(c, X_train, t_train, X_val, t_val, diff=0.001, gamma=0.1, return_ac=False):
    """
    Function fits and prints accuracy for a given diff with LinReg().
    
        args:
            c (object): Object to call accuracy on. Either LinReg or LogReg.
            X_train (numpy ndarray): Points to train/fit with.
            t_train (numpy array): Labels for X_train.
            X_val (numpy ndarray): Points to test on.
            t_val (numpy array): Labels for X_val.
            gamma (float): Learning rate.
            diff (float): Limit to when the fit-function should end. Difference in mse for each run.
            return_ac (Boolean): True if accuracy should be returned.
            
        return:
            accuracy (float:OPTIONAL): Calculated accuracy.
    """
    c.fit(X_train, t_train, gamma=gamma, diff=diff)
    accuracy = c.accuracy(X_val, t_val)
    print("Accuracy: ", accuracy)
    
    if return_ac:
        return accuracy
```

```python
# Fit and accuracy of LinReg with different values for diff

lin_reg = LinReg()

fit_and_accuracy(lin_reg, X_train, t2_train, X_val, t2_val) # diff=0.001
fit_and_accuracy(lin_reg, X_train, t2_train, X_val, t2_val, diff=0.0001)
fit_and_accuracy(lin_reg, X_train, t2_train, X_val, t2_val, diff=0.00001)
fit_and_accuracy(lin_reg, X_train, t2_train, X_val, t2_val, diff=0.000001)
fit_and_accuracy(lin_reg, X_train, t2_train, X_val, t2_val, diff=0.0000001)
fit_and_accuracy(lin_reg, X_train, t2_train, X_val, t2_val, diff=0.01)
fit_and_accuracy(lin_reg, X_train, t2_train, X_val, t2_val, diff=0.1)

"""
Total number of runs with diff as 0.001: 19 runs
Accuracy:  0.51
Total number of runs with diff as 0.0001: 51 runs
Accuracy:  0.56
Total number of runs with diff as 1e-05: 83 runs
Accuracy:  0.58
Total number of runs with diff as 1e-06: 115 runs
Accuracy:  0.6
Total number of runs with diff as 1e-07: 148 runs
Accuracy:  0.605
Total number of runs with diff as 0.01: 3 runs
Accuracy:  0.4775
Total number of runs with diff as 0.1: 2 runs
Accuracy:  0.47
"""
```

### Logistic regression

```python
def logistic(x):
    """
    Logistic function to calculate (SIGMOID)
    
        args:
            x (numpy array): values to calculate.
            
        return:
            Sigmoid(x): Sigmoid calculated on x.
    """
    return 1/(1+np.exp(-x))
```

```python
class LogReg(NumpyClassifier):

    def fit(self, X_train, t_train, gamma = 0.1, diff=0.001):
        """X_train is a Nxm matrix, N data points, m features
        t_train are the targets values for training data
        
        training on the data given.
        
        The fit function will stop when the improvement is less than 'diff' 
        
            Args:
                X_train (numpy ndarray): matrix containing points.
                t_train (numpy array): validation-list.
                gamma (float): learning rate.
                diff (float): a limit to when the fit-function should stop. Compare 'improvement' to diff.        
        
        """
        
        (k, m) = X_train.shape
        X_train = add_bias(X_train)
        
        self.theta = theta = np.zeros(m+1)
        
        # MSE
        #current = mse(X_train, t_train, theta)
        
        # Log loss
        #current = self.log_loss(X_train, t_train, theta)
        
        # Cross entropy loss
        current_c_el = self.c_el(X_train, t_train, theta)
        run_counter = 0
        
        while True:
            self.theta -= gamma / k *  X_train.T @ (self.forward(X_train) - t_train)
            #new_mse = mse(X_train, t_train, theta)
            #new_log_loss = self.log_loss(X_train, t_train, theta)
            new_c_el = self.c_el(X_train, t_train, theta)
            #diff_curr_new = current - new_mse
            #diff_curr_new = current - new_log_loss
            diff_curr_new_c_el = current_c_el - new_c_el
            
            run_counter += 1
            
            #if diff_curr_new > diff:
            if diff_curr_new_c_el > diff:
                #current = new_mse
                current_c_el = new_c_el
                #current = new_log_loss
            else:
                break
                
        print(f"Total number of runs with diff as {diff}: {run_counter} runs")
    
    def forward(self, X):
        """
        Forward move calculated with (input @ weights).
        
            args:
                X (numpy ndarray): Values to calculate.
                
            return logistic(X @ self.theta) (numpy ndarray): Dot product pushed to Sigmoid.
        """
        return logistic(X @ self.theta)
    
    def score(self, x):
        z = add_bias(x)
        score = self.forward(z)
        return score
    
    def predict(self, x, threshold=0.5):
        """
        Predicts the given values labels.
        """
        z = add_bias(x)
        score = self.forward(z)
        return (score>threshold).astype('int')
        
        
    '''def log_loss(self, X, t, theta):
        y_estimate = self.forward(X)
        #print(y_estimate)
        error = sum((-t * np.log(y_estimate)) - ((1 - t) * np.log(y_estimate)))
        #print(error)
        
        return error
        '''
        
    # Cross-entropy loss
    def c_el(self, X, t, theta):
        """
        Cross entropy loss.
        
            args:
                X (numpy ndarray): Values to calculate.
                t (numpy array): labels to X.
            
            return:
                error (numpy float): calculated error.
        """
        y_estimate = self.forward(X)
        error = -np.sum(t * np.log(y_estimate)) / y_estimate.shape[0]
        return error
```

```python
# Some imports used for this part
from collections import Counter
from operator import itemgetter
```

```python
class PyClassifier():
    """Common methods to all python classifiers --- if any"""
    
    def accuracy(self,X_test, y_test, **kwargs):
        """Calculate the accuracy of the classifier 
        using the predict method"""
        predicted = [self.predict(a, **kwargs) for a in X_test]
        equal = len([(p, g) for (p,g) in zip(predicted, y_test) if p==g])
        return equal / len(y_test)
```

```python
def distance_L2(a,b):
    """
    Euclidean distance.
    """
    return np.linalg.norm(a-b)
```

```python
def majority(a):
    """
    Finding the most common.
    """
    counts = Counter(a)
    return counts.most_common()[0][0]
```

```python
class PykNNClassifier(PyClassifier):
    """kNN classifier using pure python representations"""
    
    def __init__(self, k=3, dist=distance_L2):
        self.k = k
        self.dist = dist
        
    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train
    
    def predict(self, a):
        """
        Calculating the distance to each other point and finding 
        the k most common/closest points.
        """
        X = self.X_train
        y = self.y_train
        distances = [(self.dist(a, b), b, c) for (b, c) in zip(X, y)]
        distances.sort()
        predictors = [c for (_,_,c) in distances[0: k]]
        return majority(predictors)
```

### Simple perceptron
Finally, run the simple perceptron (week05) on the same set, and report and store accuracy.

```python
class PyPerClassifier(PyClassifier):
    """Simple perceptron python classifier"""
    
    def fit(self, X_train, y_train, eta=1, epochs=1):
        """Train the self.weights on the training data eith learning
        rate eta, running epochs many epochs
        
        training on the data given.
        
            Args:
                X_train (numpy ndarray): matrix containing points.
                t_train (numpy array): validation-list.
                eta (float): learning rate.
                ephocs (int): number of runs before ending.
        
        """
        X_train = [[1]+list(x) for x in X_train] # Put bias in position 0      
        self.dim = dim = len(X_train[0])
        self.weights = weights = [0 for _ in range(dim)]
        # Initialize all weights to 0. There are better ways!

        for e in range(epochs):
            for x, t in zip(X_train, y_train):
                y = int(self.forward(x)>0)
                for i in range(dim):
                    weights[i] += eta * (t - y) * x[i]
        
    def forward(self, x):
        """Calculate the score for the item x"""
        score = sum([self.weights[i]*x[i] for i in range(self.dim)])
        return score       
    
    def predict(self, x):
        """Predict the value for the item x"""
        x = [1] + list(x)
        score = self.forward(x)
        return int(score > 0)
```

```python
def per_acc(X_train, t_train, X_val, t_val, eta=0.1, p_range=20):
    """
    Calculating and prints the accuracy of a perceptron.
        
        Args:
            X_train (numpy ndarray): matrix containing points.
            t_train (numpy array): validation-list.
            X_val (numpy ndarray): matrix to test accuracy.
            t_val (numpy array): labels to X_val.
            eta (float): learning rate.
            p_range (int): number of runs before the perceptron stops.
        
    """
    for i in range(1,p_range):
        cl = PyPerClassifier()
        cl.fit(X_train, t_train, eta, epochs = i)
        test = cl.accuracy(X_val, t_val)
        print("Accuracy after {:2} epochs, test: {:6.3f}".format(
        i, test))
        
per_acc(X_train, t2_train, X_val, t2_val)


"""
Accuracy after  1 epochs, test:  0.647
Accuracy after  2 epochs, test:  0.650
Accuracy after  3 epochs, test:  0.593
Accuracy after  4 epochs, test:  0.588
Accuracy after  5 epochs, test:  0.647
Accuracy after  6 epochs, test:  0.625
Accuracy after  7 epochs, test:  0.583
Accuracy after  8 epochs, test:  0.650
Accuracy after  9 epochs, test:  0.650
Accuracy after 10 epochs, test:  0.595
Accuracy after 11 epochs, test:  0.583
Accuracy after 12 epochs, test:  0.565
Accuracy after 13 epochs, test:  0.613
Accuracy after 14 epochs, test:  0.647
Accuracy after 15 epochs, test:  0.593
Accuracy after 16 epochs, test:  0.583
Accuracy after 17 epochs, test:  0.662
Accuracy after 18 epochs, test:  0.598
Accuracy after 19 epochs, test:  0.595

"""
```

We saw in the lecture how a logistic regression classifier can be turned into a multi-class classifier using the one-vs-rest approach. We train one classifier for each class and assign the class which ascribes the highest probability.

Extend the logisitc regression classifier to a multi-class classifier. To do this, you must modify the target values from scalars to arrays. Train the resulting classifier on (X_train, t_train), test it on (X_val, t_val), and report the accuracy.

```python
# Logistic regression multi-class classifier.
# Target values from scalars to arrays

def scalar_to_array(scalar_list):
    """
    Turns a scalar set into an array.
    
    If there are three unique labels in the scalar_list. 
    The output array would look something like:
    [0,1,2] = [[1,0,0],[0,1,0],[0,0,1]]
    
        args:
            scalar_list(numpy array): Containing scalars (LABELS)
        
        return:
            (numpy ndarray): an array of arrays containing the size of unique labels.
    """
    # n unique labels.
    n = len(set(scalar_list))
    # Final array.
    array_t_train = []
    # Iterate over each label
    for x in scalar_list:
        array_rep = [0] * n
        array_rep[x-1] = 1
        array_t_train.append(array_rep)
    
    return np.asarray(array_t_train)
```

## Adding non-linear features

We are returning to the binary classifier and the set (X, t2). As we see, some of the classifiers are not doing too well on the (X, t2) set. It is easy to see from the plot that this data set is not well suited for linear classifiers. There are several possible options for trying to learn on such a set. The current classifiers use two features: $x_1$ and $x_2$ (and a bias term $x_0$).
```python
def three_new_features(X_train):
    """
    Takes one feature set and inserts three additional features to the given set.
    
        args:
        X_train(numpy ndarray): All the features.
        
        return:
            new_X_train(numpy ndarray): The given feature set with three additional features.
    """

    # Feature 1, 2 and 3 for each column
    x_1 = X_train[:,0] ** 2
    x_2 = X_train[:,1] ** 2
    x_3 = np.multiply(x_1,x_2)
    
    # Number of 'vectors'
    n = X_train.shape[0]
    
    # Reshape so the appending is possible
    x_1 = np.reshape(x_1,(n,1))
    x_2 = np.reshape(x_2,(n,1))
    x_3 = np.reshape(x_3,(n,1))
    
    # Adding the three columns to the feature set
    new_X_train = np.append(X_train,x_1, axis=1)
    new_X_train = np.append(new_X_train,x_2, axis=1)
    new_X_train = np.append(new_X_train,x_3, axis=1)    
    
    return new_X_train
```

```python

def normalization(X, trained=X):
    """
    Calculating normalization over a set.
    
    If you want to normalize some set with the scaled 
    values of another set, than you can use the 'trained' 
    arg.
    
        args:
            X (numpy ndarray): Values to be normalized.
            trained (numpy ndarray): A set of values that has already been normalized IF CHANGED FROM X.
    """
    # Mean value of X
    m = np.mean(trained)
    
    # Standard deviation
    s = np.std(trained)
    
    # Normalization
    X_scaled = (X - m)/s
    
    return X_scaled


print(X_train, '\n')

X_train_scaled = normalization(X_train)


print(X_train_scaled)
```

### Forwards phase
We will run the first step in the training, and start with the forward phase. Calculate the activations after the hidden layer and after the output layer. We will follow Marsland and use the logistic (sigmoid) activation function in both layers. Inspect whether the results seem reasonable with respect to format and values.

```python
# Your code


def one_forward(X_train, w, bias):
    """
    Takes a training set and it's weights and calculate the 
    activations with dot-product passed into the Sigmoid 
    function.
    
        args:
            X_train (numpy ndarray): Input-values.
            w (numpy ndarray): weights.
            bias (numpy array): bias to be added.
        
        returns:
            activations (numpy ndarray): the activations.
    
    """
    # 'ac = input @ weights + bias'
    activations = (X_train @ w) + bias

    activations = logistic(activations)
    
    return activations

```

### Backwards phase
Calculate the delta terms at the output. We assume, like Marsland, that we use sums of squared errors. (This amounts to the same as using the mean square error).

```python
# Creating the y by changing the t_train label array to an n * unique(labels) numpy array.
y_train = scalar_to_array(t_train)

# error = (Tk-Yk)Yk(1-Yk)Xi
# Y = predicted
# T = target
# X = inputs
delta_output = (y_train - output_activations) * (output_activations*(1 - output_activations))
print(delta_output)
```

Calculate the error in the hidden layer.

```python
# Calculating the hidden delta.
delta_hidden = hidden_activations * (1 - hidden_activations) * (delta_output @ w_hid_out.T)
print(delta_hidden)
```

Update the weights.
Check that they have changed.

##  Step 2: A Multi-layer neural network classifier

You want to train and test a classifier on (X, t). You could have put some parts of the code in the last step into a loop and run it through some iterations. But instead of copying code for every network we want to train, we will build a general Multi-layer neural network classfier as a class. This class will have some of the same structure as the classifiers we made for linear and logistic regression. The task consists mainly in copying in parts from what you did in step 1 into the template below. Remember to add the *self*- prefix where needed, and be careful in your use of variable names.

```python
class MNNClassifier():
    """A multi-layer neural network with one hidden layer"""
    
    def __init__(self, eta=0.001, dim_hidden=6):
        """Intialize the hyperparameters
        
            args:
                self.eta (float): learning rate.
                self.dim_hidden (float): number of nodes in the hidden layer.
        
            params:
                self.eta (int): learning rate.
                self.dim_in: number of nodes in the input layer.                
                self.dim_out: number of nodes in the output layer.
                self.b1 (numpy array): bias of -1 for w_in_hid
                self.b2 (numpy array): bias of -1 for w_hid_out
                self.y_train (numpy ndarray): labels in correct dimension.
        
        """
        self.eta = eta
        self.dim_hidden = dim_hidden

        # Should you put additional code here?
        # It's not really necessary to place the self.variables here because you could create one in the middle 
        # of the code and it will be accessable with 'self'. BUT I have added them with a 'None'-value so it's 
        # a lot easier to see which variables are accessable with the 'self'.
        
        # LAYER SIZES
        self.dim_in = None # int 
        self.dim_out = None # int 
        
        # BIASES
        self.b1 = None
        self.b2 = None
        
        # TARGETS AS AN ARRAY INSTEAD OF SCALAR
        self.y_train = None # numpy array
        
        
    def fit(self, X_train, t_train, epochs = 100):
        """Intialize the weights. Train *epochs* many epochs.
        
            args:
                X_train (numpy ndarray): matrix containing points.
                t_train (numpy array): validation-list.
                ephocs (int): number of runs before the perceptron stops.
        
        """
        
        # Initialization
        self.dim_in = X_train.shape[1]
        self.dim_out = len(set(t_train))       
        
        # Creating a y_train numpy array with shape of (m, n) where m is instances n is features
        self.y_train = scalar_to_array(t_train)
        
        # Weights, using the recommended (1/sqrt(m)) from Marsland
        self.w_in_hid = np.random.rand(self.dim_in, self.dim_hidden) * (1/np.sqrt(self.dim_in))
        self.w_hid_out = np.random.rand(self.dim_hidden, self.dim_out) * (1/np.sqrt(self.dim_in))

        # Creating the biases
        self.b1 = -np.ones(self.dim_hidden)
        self.b2 = -np.ones(self.dim_out)
        
        for e in range(epochs):
            # Run one epoch of forward-backward
            hidden_activations, output_activations = self.forward(X_train)
            
            # Calculate error with MSE derivative
            delta_output, delta_hidden = self.mse_MLP(output_activations, hidden_activations)
            
            # Update weights
            self.update_weights(X_train, delta_hidden, delta_output, hidden_activations)
        
        
    def forward(self, X):
        """Perform one forward step. 
        Return a pair consisting of the outputs of the hidden_layer
        and the outputs on the final layer
        
        
            args:
                X (numpy ndarray): the input values.
                
            returns:
                hidden_activations (numpy ndarray): The activation-matrix of hidden.
                output_activations (numpy ndarray): The activation-matrix of output.
        """
        # Calculating the activations
        # X is the input to the weights
        # calculation: (X @ w1) + the bias
        hidden_activations = (X @ self.w_in_hid) + self.b1
        # Sigmoid
        hidden_activations = logistic(hidden_activations)
        
        # the hidden activations is the input of the weights
        output_activations = (hidden_activations @ self.w_hid_out) + self.b2
        # Sigmoid
        output_activations = logistic(output_activations)
        
        
        return hidden_activations, output_activations
    
    
    def mse_MLP(self, output_activations, hidden_activations):
        """
        Calculating the mse on the output-layer and then the hidden-layer.
        
            args:
                output_activations (numpy ndarray): the calculated hidden activations.
                hidden_activations (numpy ndarray): the calculated hidden activations.
                
            returns:
                delta_output (numpy ndarray): error of the outputlayer.
                delta_hidden (numpy ndarray): error of the hidden layer.
        """
        # As in Marshland, using the derivative.
        delta_output = (self.y_train - output_activations) * (output_activations*(1 - output_activations))
        delta_hidden = hidden_activations * (1 - hidden_activations) * (delta_output @ self.w_hid_out.T)
        
        return delta_output, delta_hidden
    
    def update_weights(self, X_train, delta_hidden, delta_output, hidden_activations):
        """
        Updating the weights in the MLP system.
        Adding the calculated change to the weights.
        
            args:
                X_train (numpy ndarray): matrix containing points.
                delta_hidden (numpy ndarray): error of the hidden layer.
                delta_output (numpy ndarray): error of the outputlayer.
                hidden_activations (numpy ndarray): the calculated hidden activations.
        """
        # Apply the changes with the learning rate.
        update_w_in_hid = self.eta * (X_train.T @ delta_hidden)
        update_w_hid_out = self.eta * (hidden_activations.T @ delta_output)
        
        # Add changes to our existing weights
        self.w_in_hid += update_w_in_hid
        self.w_hid_out += update_w_hid_out
    
    def accuracy(self, X_test, t_test):
        """Calculate the accuracy of the classifier on the pair (X_test, t_test)
        Return the accuracy.
            
            args:
                X_test (numpy ndarray): matrix containing points we want to test.
                t_test (numpy array): Labels to X_test.
                
            return:
                accuracy (float): accuracy of the predicted outcome.
        """
        # Converting target to the right dimensions
        target = scalar_to_array(t_test)
        
        # Predicted results
        predicted = self.predict(X_test)
        
        # For each row value becomes 1, rest to 0.
        pred_formatted = np.zeros_like(predicted)
        pred_formatted[np.arange(len(predicted)), predicted.argmax(1)] = 1
        
        n, m = pred_formatted.shape
    
        tp = 0
        # Comparing each row, finding the rows that are equal.
        for x in range(n):
            if np.array_equal(pred_formatted[x], target[x]):
                 tp += 1

        accuracy = tp/n
        
        return accuracy
        
    
    def predict(self, X):
        """
        Takes a set of values and calculates the 
        predicted outcome with a 'forward move'.
        Returning the 'output-layer'.
        
            args:
                X (numpy ndarray): set of values/points to be predicted.
            
            return:
                predicted (numpy ndarray): predicted values.
        """
        hidden, predicted = self.forward(X)
        
        return predicted
```

```python
def confusion_matrix(target, predicted):
    """
    Calculating confusion matrix.
    This will give the same result as scikit.
    
        args:
            target (numpy ndarray): a matrix of the targets.
            predicted (numpy ndarray): a matrix of the predicted values.
            
        return:
            c_m (numpy ndarray): The confusion matrix of the given targets and predicted.
    """
    target = scalar_to_array(target)
    
    pred_formatted = np.zeros_like(predicted)
    pred_formatted[np.arange(len(predicted)), predicted.argmax(1)] = 1
    
    target = target.argmax(axis=1)
    predicted = pred_formatted.argmax(axis=1)
    
    # The number of classes. Need to create right size of matrix.
    n = len(np.unique(target))
    
    # Confusion matrix initiated with zeros.
    c_m = np.zeros((n, n))
    
    # Loop over the values
    for i in range(len(target)):
        c_m[target[i]][predicted[i]] += 1
    
    return c_m
```

```python
def precision_and_recall(confusion_matrix):
    """
    Calculating the precision and recall with 
    the use of a given confusion matrix.
    
        args:
            confusion_matrix (numpy ndarray): Confusion matrix to be used.
        
        returns:
            precisions (list): Containing the calculated precisions for all classes.
            recalls (list): Containing the calculated recalls for all classes.
    """
    precisions = []
    recalls = []
    
    # precision
    class_index = 0
    for x in confusion_matrix.T:
        prec = x[class_index]/sum(x)
        precisions.append(prec)
        class_index += 1
    
    # recall
    class_index = 0
    for x in confusion_matrix:
        rec = x[class_index]/sum(x)
        recalls.append(rec)
        class_index += 1
        
        
    return precisions, recalls
```

```python
# MLP ACCURACY, PRECISION AND RECALL

mnn = MNNClassifier(learning_rate)

mnn.fit(X_train_scaled, t_train, ephocs)

print("accuracy (X_train, t_train, X_test, t_test): ", mnn.accuracy(X_test_scaled, t_test))

predict_test = mnn.predict(X_test_scaled)
mnn_confusion_matrix = confusion_matrix(t_test ,predict_test)
print(mnn_confusion_matrix)
precision, recall = precision_and_recall(mnn_confusion_matrix)
print("precision:", precision, "\nrecall:", recall)
```
